```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com")
```

```python
#Let's first close the cookie window by click the "no thanks" button
button = driver.find_element('id','W0wltc')
button.click()
```

```python
# Find the search bar using the name of the input field
search_bar = driver.find_element("name", "q")

# Type the search term and hit ENTER
search_bar.send_keys("roaddanger")
```

```python
#Click enter!
search_bar.send_keys(Keys.RETURN)

# Wait for some time to let the results load
time.sleep(2) 

#The page will now have made the search!
```

```python
# click on the first link
results = driver.find_elements(by=By.CSS_SELECTOR,value='a h3') #Select all links under the div with id search: these are the search results.
results[0].click()
```

```python
buttons = driver.find_elements(by=By.CSS_SELECTOR,value='.button')

for button in buttons:
    if 'Akkoord' in button.text:
        button.click()
```

```python
input_field = driver.find_element(By.ID, 'searchText')

# Send text to the input field
input_field.send_keys('*fiets*')
```

```python
buttons = driver.find_elements(by=By.CSS_SELECTOR,value='.button')

for button in buttons:
    if 'Zoek' in button.text:
        button.click()
```

```python
# Find the scrollable container
scrollable_div = driver.find_element(By.CSS_SELECTOR, "#pageMain")

# Scroll to the bottom of the container
for i in range(1000):
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scrollable_div)
```

```python
# Running the same code as above, we now get a larger number of results!
results = driver.find_elements(by=By.CSS_SELECTOR,value='.cardTitle') #Select all links under the div with id search: these are the search results.

# Extract and print the top 10 hits 
i=0
for result in results:
    if len(result.text)>0:
        i+=1
        print(f"{result.text}")
        print(i)
```

```python
# Running the same code as above, we now get a larger number of results!
results = driver.find_elements(by=By.CSS_SELECTOR,value='.cardArticle') #Select all links under the div with id search: these are the search results.
links = []

# Loop through each article div and find the <a> tag within it
for article in results:
    try:
        # Find the <a> tag within the current div
        link_element = article.find_element(By.TAG_NAME, "a")
        
        # Print the href attribute of the <a> tag
        print(link_element.get_attribute("href"))
        links.append(link_element.get_attribute("href"))

    except:
        # If no <a> tag is found, handle it (optional)
        print("No link found in this cardArticle")
```

```python
# Locate all elements with class `crashCardList`
cards = driver.find_elements(By.CSS_SELECTOR, '.cardCrashList')

# Loop through each card to extract the title and URL
for card in cards:
    try:
        # Find the title inside each `crashCardList`
        title_element = card.find_element(By.CLASS_NAME, "cardTitle")
        title_text = title_element.text if title_element else "No Title Found"
        
        # Find the link inside `cardArticle` and get its href
        link_element = card.find_element(By.CSS_SELECTOR, ".cardArticle a")
        link_url = link_element.get_attribute("href") if link_element else "No URL Found"
        
        # Print the title and URL
        print(f"Title: {title_text}")
        print(f"URL: {link_url}")
        print("-" * 30)  # Separator for readability
        print(f"Number of results: {len(cards)}")
    except Exception as e:
        # Handle cases where title or link might be missing
        print("Title or URL not found for a card:", e)
```

```python
driver.quit()
```

```python
# click on the first link
results = driver.find_elements(by=By.CSS_SELECTOR,value='a h3') #Select all links under the div with id search: these are the search results.
results[0].click()
```

```python
input_element = driver.find_element(By.CSS_SELECTOR, 'input#url')
articles = []
# Send text to the input field
for link in links:
    # For now only using articles from DPG Media and NRC because they have the same layout
    websites = ['https://www.ad.nl/', 'https://www.gelderlander.nl/', 'https://www.ed.nl/', 'https://www.volkskrant.nl/', 'https://www.nrc.nl/', 'https://www.trouw.nl/', 'https://www.parool.nl/', 'https://www.bndestem.nl/', 'https://bd.nl/', 'https://www.tubantia.nl/', 'https://www.pzc.nl/', 'https://www.destentor.nl/']
    if any(website in link for website in websites):
        input_element.send_keys(link)
        input_element.send_keys(Keys.RETURN)
        # Sometimes a popup appears, we need to close it
        try:
            close_btn_element = driver.find_element(By.CSS_SELECTOR, '#ALREADY_CLOSEBTN')
            close_btn_element.click()
        except:
            pass
        # Articles that have never been requested before need to be loaded first
        article_element = WebDriverWait(driver, 900).until(
        EC.visibility_of_element_located((By.TAG_NAME, 'article'))
        )
        # Get the data from the article
        article_element = driver.find_element(By.TAG_NAME, 'article')
        article_text = article_element.text
        articles.append(article_text)
        print(article_text)
        # Wait for some time and reload the page to clear the input field
        time.sleep(2)
        driver.get("https://archive.ph/")
        input_element = driver.find_element(By.CSS_SELECTOR, 'input#url')
        input_element.click()
        input_element.clear()
        
file_name = "bike_articles.csv"

# Write to CSV file
with open(file_name, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    
    # Write each article on a new line
    for article in articles:
        writer.writerow([article])

print(f"Articles saved to {file_name}")
```
